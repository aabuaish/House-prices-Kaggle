```python
def dataConverter(dftrain,dftest):
    # filling missing data in the lonfrontage with the average value
    ind=dftrain[dftrain['LotFrontage'].isnull()].index
    dftrain.loc[ind,'LotFrontage']=dftrain['LotFrontage'].mean()

    ind=dftest[dftest['LotFrontage'].isnull()].index
    dftest.loc[ind,'LotFrontage']=dftest['LotFrontage'].mean()
    
    # dropping highly missing datas
    dftrain.drop(['Alley','PoolQC','Fence','MiscFeature','FireplaceQu'],axis=1,inplace=True)

    dftest.drop(['Alley','PoolQC','Fence','MiscFeature','FireplaceQu'],axis=1,inplace=True)
    dftest['SalePrice']=0
    
    # filling missing data in the test set with most common values for categorical 
    # features and with mean for numerical features
    tmpD = dftest.isnull().sum(axis=0)
    cols = list(tmpD[tmpD>0].index)
    dic = {}

    for var in cols:
        if type(dftest[var][0]) is str:
            dic[var] = (dftest[var].value_counts().index[0],dftest[var].value_counts().max())
            dftest.loc[:,var] = dic[var][0]
        else:
            dftest.loc[:,var]=dftrain[var].mean()
            
    
    # concatenating dataframes
    tmp = pd.concat([dftrain,dftest],axis=0,join='inner')
    tmp.reset_index(drop=True,inplace=True)
    
    # droping missing data in the training set
    tmp.dropna(inplace=True)
    
    # one_hot conversion
    categories=pd.get_dummies(tmp.select_dtypes(exclude=['floating','integer']))
    X = pd.concat([tmp.select_dtypes(include=['floating','integer']),categories],axis=1)
    
    # splitting the combined dataframe
    ind = X['Id']>(dftrain.iloc[-1]['Id'])
    
    # getting Xtrain and Xtest
    Xtrain=X[~ind]
    Xtrain = Xtrain.reset_index(drop=True)
    Xtrain.drop('Id',inplace=True,axis=1)
    
    Xtest=X[ind]
    Xtest = Xtest.reset_index(drop=True)
    testId = pd.DataFrame(Xtest['Id']) # grapping the test set Ids
    Xtest.drop('Id',inplace=True,axis=1)
    
    # getting the labels for the training set
    Y = pd.DataFrame(Xtrain.loc[:,'SalePrice'])
    
    # dropping the price from the features variables
    Xtrain.drop('SalePrice',inplace=True,axis=1)
    Xtest.drop('SalePrice',inplace=True,axis=1)
    
    return (Xtrain,Xtest,Y,testId)
```

```python
dftrain = pd.read_csv('train.csv')
dftest = pd.read_csv('test.csv')
(XtrainD,XtestD,YD,IdD)=dataConverter(dftrain,dftest)
```
