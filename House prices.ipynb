```python
def dataConverter(df):
    # droping columns with highly missing data
    dftrain2 = df.drop(['Id','Alley','PoolQC','Fence','MiscFeature','FireplaceQu'],axis=1)
    
    # filling missing data with average
    ind=dftrain2[dftrain2['LotFrontage'].isnull()].index
    dftrain3 = dftrain2.copy()
    dftrain3.loc[ind,'LotFrontage']=dftrain2['LotFrontage'].mean()
    
    # droping missing data in catagorical features
    dftrain3.dropna(inplace=True)
    
    # performing one_hot catogrization
    categories=pd.get_dummies(dftrain3.select_dtypes(exclude=['floating','integer']))
    if 'SalePrice' in df.columns: 
        dftrain4 = dftrain3.select_dtypes(include=['floating','integer']).drop('SalePrice',axis=1)
    else:
        dftrain4 = dftrain3.select_dtypes(include=['floating','integer'])
    
    dftrain4 = pd.concat([dftrain4,categories],axis=1)
    
    # getting the labels dataframe
    if 'SalePrice' in df.columns: 
        Y = pd.DataFrame(dftrain3['SalePrice'])
    else:
        Y = None
    
    # copying the features dataframe with one_hot catogrization into the variable X_train
    X = dftrain4.copy()
    
    return (X,Y)
```

```python
(x_train,Y_train)=dataConverter(dftrain)
```
